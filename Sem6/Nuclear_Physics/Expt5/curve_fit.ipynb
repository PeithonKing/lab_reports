```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
```

## Defining the expected function

```python
def func(x, A, B):
    return A/(np.sin(0.5*0.0174532925199*(x-B))**4)
```

```python
xs = np.array([25, 20, 15, 10, 5, -5, -10, -15, -20, -25])
ys = np.array([0.225, 0.940, 2.929, 17.626, 29.286, 29.343, 17.824, 2.940, 1.009, 0.225])
```

```python
# popt, pcov = curve_fit(func, xs, ys)
popt = [0.000109843, -0.00126324]
```

## Showing Results

```python
xFit = np.arange(-30, -4.7, 1)
plt.plot(xs, ys, 'go')
plt.plot(xFit, func(xFit, *popt), 'r')
plt.plot(-xFit, func(-xFit, *popt), 'r')

text = f"""
A = {popt[0]} per second
B = {popt[1]} degree
"""
plt.text(0.15, 0.6, text, fontsize=12, transform=plt.gcf().transFigure)
plt.xlabel(r'Angle $(^\circ)$')
plt.ylabel(r'N (counts per sec)')
```
