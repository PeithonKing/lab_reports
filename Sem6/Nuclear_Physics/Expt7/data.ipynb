```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def func(x, m, c):
    return m*x + c
```

```python
a = np.array([0.1 , 0.5, 1 , 1.5, 2 , 2.5, 3 , 3.5, 4 , 4.5, 5])
a
```

```python
{i/51:"0"*(8-len(bin(int(i))[2:]))+bin(int(i))[2:] for i in a*51}
```

```python
b = ["0"*(8-len(bin(int(i))[2:]))+bin(int(i))[2:] for i in a*51]
```

```python
y = [round(int(i.lstrip("0"), 2)/51, 4) for i in b]
```

```python
popt, pcov = curve_fit(func, a, y)
```

```python
popt
```

```python
plt.plot(a, func(a, *popt), label="fitted curve")


plt.plot(a, y, "ro", label="data")
plt.xlabel("Input (V)")
plt.ylabel("Output (V)")

text = f"""
fited in y = mx + c
m = {round(popt[0], 6)}
c = {round(popt[1], 6)}
"""
plt.text(0.3, 0.5, text, fontsize=12, transform=plt.gcf().transFigure)
plt.legend()
```
