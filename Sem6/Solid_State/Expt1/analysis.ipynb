```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
```

```python
def func(x, m, c):  # this is a 'straight line'
    return m*x + c
```

### Germanium

```python
xs = np.array([0.00, 0.21, 0.38, 0.43, 0.60, 0.80, 1.06, 1.25, 1.42, 1.60, 1.83, 2.02, 2.16, 2.40, 2.63, 2.85, 3.05, 3.23, 3.53, 3.68, 3.82, 3.92, 4.02, 4.31, 4.54, 4.80, 5.03, 5.12, 5.38, 5.65, 5.98, 6.23, 6.51, 6.60, 7.01, 7.31])/1000
ys = np.array([0, 0.019, 0.034, 0.038, 0.054, 0.071, 0.094, 0.11, 0.124, 0.142, 0.162, 0.178, 0.191, 0.211, 0.232, 0.25, 0.268, 0.285, 0.311, 0.324, 0.337, 0.345, 0.354, 0.38, 0.399, 0.423, 0.442, 0.451, 0.474, 0.497, 0.527, 0.548, 0.573, 0.582, 0.616, 0.643])
```

```python
popt, pcov = curve_fit(func, xs, ys)
```

```python
xFit = np.arange(-0.0001, 0.0075, 0.0001)
plt.plot(xs, ys, 'ko', label='Data')
plt.plot(xFit, func(xFit, *popt), 'k', label='Fit')

text = rf"""
m = ${popt[0]:.2f} \pm {pcov[0,0]**0.5:.2f}$ohm
c = ${popt[1]:.2f} \pm {pcov[1,1]**0.5:.2f}V$
"""
plt.text(0.4, 0.1, text, fontsize=20, transform=plt.gcf().transFigure)
plt.xlabel(r'Current (I) in A')
plt.ylabel(r'Potential Diiference (V) in V')
plt.legend()
```
