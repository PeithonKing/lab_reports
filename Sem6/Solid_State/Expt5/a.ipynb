```python
import matplotlib.pyplot as plt
```

```python
x = [9, 18, 27, 36, 45, 54, 63, 72, 81, 90, 99, 108, 117, 126, 135]
y = [1.8, 7, 12.1, 15.4, 19.5, 21.8, 27.07, 30.5, 34, 38.8, 42, 48, 57, 63, 77]
```

```python
plt.plot(x, y)
plt.plot(x, y, "ko")
plt.xlabel(r"$\theta (^\circ)$")
plt.ylabel(r"Frequency ($\nu$) (KHz)")
```

```python
x = [18, 36, 54, 72, 90, 108, 126, 144, 162, 180, 198, 216]
y = [9.206, 11.706, 13.836, 16.136, 18.246, 20.246, 21.966, 23.868, 25.768, 26.968, 40.668, 41.768]
```

```python
n = 2
plt.plot(x[:-n], y[:-n], "#237ab5", label = "Acoustical Branch")
plt.plot(x[:-n], y[:-n], "#237ab5", marker="o")
plt.plot(x[-n:], y[-n:], "#ff983d", label = "Optical Branch")
plt.plot(x[-n:], y[-n:], "#ff983d", marker="o")
plt.xlabel(r"$\theta (^\circ)$")
plt.ylabel(r"Frequency ($\nu$) (KHz)")
plt.legend()
```
