```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x, a, b, c):
	return a - c*np.e**(b * x)
```

## pink_inside

```python
data_points = 17
l = 0.17
h = 0.731

a = 893.565
b = 2.23186
c = 172.822

x = np.linspace(h, l, data_points) + np.random.normal(0, 0.0005, data_points)
y = func(x, a, b, c) + np.random.normal(0, 20, data_points)
plt.plot(x, y, 'o')
```
